# Graphs of trainin results

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## MSE Loss

```python
df=pd.read_csv('tables_training/trainingVSTest_b1xp0_edit.csv')
print(df)
b1xp0=df.to_numpy()
print(type(df))
print(type(b1xp0))
```

```python
fig = plt.figure(figsize=(12,7.5))
plt.plot(b1xp0[:,0],b1xp0[:,1],label='Training')
plt.plot(b1xp0[:,0],b1xp0[:,2],'--',label='Validation')
plt.ylabel('MSELoss')
plt.xlabel('Epochs')
plt.legend(loc='upper right', fontsize=12, frameon=True)

a=plt.axes([0.35,0.35,0.5,0.3125])
plt.plot(b1xp0[:,0],b1xp0[:,1],label='Training')
plt.plot(b1xp0[:,0],b1xp0[:,2],'--',label='Validation')
plt.grid('on')
plt.ylim(0.005,0.008)

plt.show()
```
